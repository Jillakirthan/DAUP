```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#dataset in url type
file_path = "/content/insurance.csv"
data = pd.read_csv(file_path)
# url = "https://www.kaggle.com/datasets/awaiskaggler/insurance-csv"
# data = pd.read_csv(url)
#you can change the charges
print(data.head(50))
```

a) Create vectors, factors, lists, matrices, and data frames.

```python
#Vector
age_vector = data['age']
print("Vector of ages:", age_vector[:5])

#Factor
smoker_factor = data['smoker'].astype('category')
print("Factor of smoker:", smoker_factor.cat.categories)

#List
my_list = [age_vector, smoker_factor, data['expenses']]
print("List of elements:", my_list[:2])

#Matrix
matrix = data[['age', 'bmi', 'expenses']].to_numpy()
print("Matrix:\n", matrix[:5])

#DataFrame
print("Data Frame:\n", data.head())
```

b) Generate a scatter plot. Customize it by adding a title, axis labels, text annotations, and changing the plot characters (e.g., circles to plus signs).

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=data['age'], y=data['expenses'], hue=data['smoker'], style=data['sex'])
plt.title("Scatter Plot: Age vs Charges (Smoking Status)")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.text(25, 50000, "High expenses for smokers", fontsize=12, color='red')
plt.legend(title='Smoker')
plt.grid(True)
plt.show()
```

c) Filter the data based on a condition and updates the plot to reflect the filtered data.

Plot a scatter plot between age(20 to 40) and smoking,

```python
filtered_data = data[(data['age'] >= 20) & (data['age'] <= 40)]

plt.figure(figsize=(10, 5))

sns.scatterplot(x=filtered_data['age'], y=filtered_data['expenses'], hue=filtered_data['smoker'], style=filtered_data['sex'])
plt.title("Scatter Plot: Age (20-40) vs Charges (Smoking Status)")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.legend(title='Smoker')
plt.grid(True)
plt.show()
```

Find expenses of smokers and non smokers,(plot)

```python
expenses = data.groupby('smoker')['expenses'].sum()
print("Expenses by Smoking Status:\n", expenses)

expenses.plot(kind='bar', color=['red', 'blue'], figsize=(8, 5))
plt.title("Total Expenses by Smoking Status")
plt.ylabel("Total Charges")
plt.xlabel("Smoker Status")
plt.xticks(rotation=0)
plt.show()
```

Find the number of smokers gender wise (plot)

```python
smokers_gender = data[data['smoker'] == 'yes'].groupby('sex')['smoker'].count()
print("Number of Smokers by Gender:\n", smokers_gender)

smokers_gender.plot(kind='bar', color=['green', 'blue'], figsize=(8, 5))
plt.title("Number of Smokers by Gender")
plt.ylabel("Number of Smokers")
plt.xlabel("Gender")
plt.xticks(rotation=0)
plt.show()
```

Plot a scatter plot between age and expenses,

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=data['age'], y=data['expenses'], hue=data['smoker'], style=data['sex'])
plt.title("Scatter Plot: Age vs Expenses")
plt.xlabel("Age")
plt.ylabel("Expenses")
plt.legend(title='Smoker')
plt.grid(True)
plt.show()
```

Find the number of teenagers ages 13 to 19

```python
teenagers = data[(data['age'] >= 13) & (data['age'] <= 19) & (data['smoker']=='yes')]
number_of_teenagers = len(teenagers)
print(f"Number of teenage members: {number_of_teenagers}")
```
